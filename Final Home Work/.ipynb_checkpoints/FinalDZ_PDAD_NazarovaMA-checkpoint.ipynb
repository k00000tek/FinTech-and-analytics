### Задание
Есть мнение: раньше снимали настоящее кино, не то что сейчас. Ваша задача — проверить это утверждение, используя файлы с рейтингами фильмов movies_stats.csv. То есть проверить, верно ли, что с ростом года выпуска фильма средний рейтинг становится ниже.

```python
import pandas as pd
```

```python
# Создаем список из всех годов с 1950-го по 2010-ый (в виде строк)
years = [str(i) for i in range(1950, 2011)]
```

```python
def production_year(row_title, years_list):
    """
    Возвращает год выпуска фильма по строке его названия.
    
    production_year('Dangerous Minds (1995)')
    1995
    """
    for year in years_list:
        if year in row_title:
            # если в строке названия фильма найдена подстрока из списка годов, то возвращается год выпуска (числ.)
            return int(year)
    # если в строке названия фильма подстрока из списка годов НЕ найдена, то возвращается 1900-ый год
    return 1900
```

```python
# Импортируем таблицу рейтинга фильмов
movies_stats_df = pd.read_csv('movies_stats.csv')
movies_stats_df.head()
```

```python
# Каждому фильму в столбец years записываем год выпуска
movies_stats_df['year'] = movies_stats_df['title'].apply(production_year, years_list=years)
movies_stats_df.head()
```

```python
# Вычисляем среднее значение рейтинга по годам и сортируем по убыванию среднего рейтинга
rating_by_year = movies_stats_df.groupby('year').agg({'rating':'mean'}).sort_values('rating', ascending=False)
# В результате получаем столбец среднего рейтинга с индексами, равными году выпуска фильмов
rating_by_year
```

### Результат
Можно наблюдать, что, действительно, с ростом года выпсука фильма средний рейтинг, хоть и немонотонно, но снижается.

```python
# Для иллюстрации выявленной тенденции, можно построить график
plot = rating_by_year.plot(title="Movies rating by years")
```
